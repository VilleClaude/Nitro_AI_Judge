# Read data


```python
import pandas as pd
```

```python
train = pd.read_csv('/content/train_fitness_data.csv')
test = pd.read_csv('/content/test_fitness_data.csv')
sample_output = pd.read_csv('/content/sample_output_fitness.csv')
```

```python
train.head(5)
```

# Subtask 1


```python
train['BMI'] = train['weight_kg'] / ((train['height_cm'] / 100)**2)
test['BMI'] = test['weight_kg'] / ((test['height_cm'] / 100)**2)

display(train.head())
```

```python
test.head(5)
```

```python
test['label_subtask1'] = (test['BMI'] > 25).astype(int)
output_subtask1 = test[['ID', 'label_subtask1']]
display(output_subtask1.head())
```

# Subtask 2


```python
train = train.drop('ID', axis=1)
display(train.head())
```

```python
train.head(100)
```

```python

train['gender'] = train['gender'].replace({'M': 0, 'F': 1}).infer_objects(copy=False)
display(train['gender'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
display(train['age'].describe())
plt.figure(figsize=(10, 6))
sns.histplot(train['age'], kde=True)
plt.title('Distribution of Age in Training Data')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

train['smokes'] = train['smokes'].replace({'yes': 1, 'smoke': 1, 1: 1, 'no': 0, 9: 0}).astype(int)
display(train['smokes'].value_counts())
```

```python
train.head(5)
```

```python
train = train.fillna(train.mean())
display(train.info())
```

```python
train = train[(train['BMI'] >= 15) & (train['BMI'] <= 35)]
print(f"Shape of the train dataframe after removing rows with BMI outside 15-35: {train.shape}")
```

```python
train_x = train.drop('is_fit', axis=1)
train_y = train['is_fit']
display(train_x.head())
display(train_y.head())
```

```python
from sklearn.model_selection import train_test_split
train_x, val_x, train_y, val_y = train_test_split(train_x, train_y, train_size=0.8, random_state=42)
print("Shape of train_x:", train_x.shape)
print("Shape of val_x:", val_x.shape)
print("Shape of train_y:", train_y.shape)
print("Shape of val_y:", val_y.shape)
```

## Random Forest Model


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(train_x, train_y)
val_y_pred_random_forest = rf_model.predict(val_x)

accuracy = accuracy_score(val_y, val_y_pred_random_forest)
print(f"Accuracy on the validation set: {accuracy}")
```

```python
from sklearn.metrics import f1_score
f1 = f1_score(val_y, val_y_pred_random_forest)
print(f"F1 Score on the validation set: {f1}")
```

## Submit for random forest

```python

test['gender'] = test['gender'].replace({'M': 0, 'F': 1}).infer_objects(copy=False)
test['smokes'] = test['smokes'].replace({'yes': 1, 'smoke': 1, 1: 1, 'no': 0, 9: 0}).astype(int)

test = test.fillna(test.mean())

test_x = test[train_x.columns]

test_y_pred_random_forest = rf_model.predict(test_x)

output_subtask2 = pd.DataFrame({'ID': test['ID'], 'label_subtask2': test_y_pred_random_forest})

output_subtask1_long = output_subtask1.melt(id_vars='ID', var_name='subtaskID', value_name='answer')
output_subtask1_long['subtaskID'] = output_subtask1_long['subtaskID'].replace({'label_subtask1': 1})

output_subtask2_long = output_subtask2.melt(id_vars='ID', var_name='subtaskID', value_name='answer')
output_subtask2_long['subtaskID'] = output_subtask2_long['subtaskID'].replace({'label_subtask2': 2})


final_output = pd.concat([output_subtask1_long, output_subtask2_long], ignore_index=True)

final_output = final_output.rename(columns={'ID': 'datapointID'})

final_output = final_output.sort_values(by=['datapointID', 'subtaskID']).reset_index(drop=True)


display(final_output.head())

final_output.to_csv('submission.csv', index=False)
```
